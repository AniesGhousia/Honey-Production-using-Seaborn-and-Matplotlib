```python
# Calculate the total production for each year
total_production = honey_df.groupby('year')['totalprod'].sum()
total_production
```

```python
# Calculate the percentage distribution of total production for each year
percent_distribution = total_production / total_production.sum() * 100
percent_distribution
```

```python
percent_distribution.index
```

```python
# Create a pie chart
plt.pie(percent_distribution, labels = percent_distribution.index, autopct='%1.1f%%')
plt.title('\nPercentage Distribution of Total Production by Year\n\n')
plt.axis('equal')  
plt.show()
```

```python
# To create a correlation matrix

columns = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']
correlation_matrix = honey_df[columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='inferno', vmin=-1, vmax=1)
plt.title('Correlation Between Variables')
plt.show()
```

<html>
    <head>
        <body>
            <h3><u><font color = green> Observations :</font></u></h3><br>
            <li>The heatmap plot visualizes the correlation between the selected columns using  <b><i><font color=#C1063C>color-coded cells.</font></i></b>  Each cell represents the correlation coefficient between two variables.</li><br>
<li>Positive correlations are indicated by  <b><i><font color=#C1063C>warmer colors </font></i></b>(e.g., red), while negative correlations are indicated by  <b><i><font color=#C1063C>cooler colors</font></i></b> (e.g., blue).</li><br>
<li>The intensity of the color represents the  <b><i><font color=#C1063C>strength of the correlation,</font></i></b> with darker shades indicating stronger correlations.</li><br>
<li>You can observe the patterns and relationships between the variables based on the  <b><i><font color=#C1063C>color distribution </font></i></b>in the heatmap.</li><br>
<li>Correlation values close to 1 or -1 indicate a  <b><i><font color=#C1063C>strong positive or negative linear </font></i></b>relationship, respectively. Values close to 0 indicate a  <b><i><font color=#C1063C>weak or no linear relationship.</font></i></b></li><br>
<li>The diagonal line of the heatmap represents the correlation of each variable with itself, which is always <b><i><font color=#C1063C> 1 (or 100%).</font></i></b></li><br>
<li>You can use the correlation values to identify which variables are  <b><i><font color=#C1063C>strongly correlated or inversely correlated </font></i></b>with each other.</li><br>
<li>The correlation matrix provides insights into the interdependencies between the selected variables, helping  <b><i><font color=#C1063C>identify potential associations</font></i></b> and <b><i><font color=#C1063C> guiding further analysis.</font></i></b></li><br>
